```python
df = pd.read_csv('clean_inventario.csv')
df.to_sql('invt_sql', conn, index=False, if_exists='replace')
```

```python
df.columns = df.columns.str.strip()
```

```python
category_summary = df.groupby('category').agg({
    'units_sold': 'sum',
    'inventory_level': 'mean',
    'demand_forecast': 'mean'
}).reset_index()
print(category_summary)

print("\n--- MÉTRICAS POR REGIÓN ---")
region_summary = df.groupby('region').agg({
    'units_sold': 'sum'
}).reset_index()
print(region_summary)

pivot_sales = df.pivot_table(index='category', columns='region', values='units_sold', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Ventas Totales por Categoría y Región")
plt.ylabel("Categoría de Producto")
plt.xlabel("Región de la Tienda")
plt.show()
```

Realizar el análisis exploratorio de datos (EDA)

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)


print("--- ESTRUCTURA E INFORMACIÓN DEL DATASET ---")
print(df.info())

print("\n--- VALORES NULOS DETECTADOS ---")
print(df.isnull().sum())


```

```python

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

sns.countplot(data=df, x='category', order=df['category'].value_counts().index, ax=axes[0], palette='Blues_r')
axes[0].set_title('Distribución de Registros por Categoría de Producto')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)

sns.countplot(data=df, x='region', order=df['region'].value_counts().index, ax=axes[1], palette='Oranges_r')
axes[1].set_title('Distribución de Tiendas por Región Geográfica')

plt.tight_layout()
plt.show()
```

```python


print("\n--- MÉTRICAS DE VENTAS E INVENTARIO POR CATEGORÍA ---")
category_analysis = df.groupby('category').agg(
    Ventas_Totales=('units_sold', 'sum'),
    Ventas_Promedio=('units_sold', 'mean'),
    Inventario_Promedio=('inventory_level', 'mean'),
    Pronostico_Demanda_Promedio=('demand_forecast', 'mean')
).reset_index()
print(category_analysis.to_string(index=False))

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='category', y='units_sold', palette='Set2')
plt.title('Dispersión y Outliers de Unidades Vendidas por Categoría')
plt.xticks(rotation=30)
plt.show()
```

```python

pivot_sales = df.pivot_table(
    index='category',
    columns='region',
    values='units_sold',
    aggfunc='sum'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, annot=True, fmt=",d", cmap="YlGnBu", linewidths=.5)
plt.title('Mapa de Calor: Unidades Vendidas por Categoría y Región')
plt.ylabel('Categoría de Producto')
plt.xlabel('Región de la Tienda')
plt.show()
```

```python

pivot_weather = df.pivot_table(
    index='region',
    columns='weather_condition',
    values='units_sold',
    aggfunc='mean'
)

plt.figure(figsize=(10, 5))
sns.heatmap(pivot_weather, annot=True, fmt=".1f", cmap="magma", linewidths=.5)
plt.title('Ventas Promedio Diario por Región y Condición Climática')
plt.ylabel('Región')
plt.xlabel('Condición Climática')
plt.show()
```
