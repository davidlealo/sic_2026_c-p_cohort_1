# **EXTRACCION DE DATOS**

Para el presente proyecto, nuestros datos crudos fueron extraidos mediante la API pública entregada por la Comisión Nacional de Energía.

* ### **Fuente:** Comision Nacional de Energía (https://apidocs.cne.cl/)
* ### **Recursos:**
  - #### Lista Estaciones de Servicio (https://api.cne.cl/api/v4/estaciones)
  - #### Lista tipos de combustibles (https://api.cne.cl/api/v4/combustible/vehicular/distribuidores)
  - #### Lista Distribuidores Combustible Vehicular (https://api.cne.cl/api/v4/combustible/vehicular/distribuidores)

Utilizamos como base la información de más de 2000 bencineras a lo largo de todo **Chile**, de las cuales cubrian datos como **digitos identificadores** para cada bomba, **horario de atencion**, **marca de distribuidores**, si posee **servicios** como **baño o tienda**, **precios por tipo de combustible** y sus **unidades de cobro**, etc. La base de datos precisamente abarca las siguientes regiones:


* Región de Tarapacá


* Región de Antofagasta


* Región de Atacama


* Región de Coquimbo


* Región de Valparaíso


* Región del Libertador Gral. Bernardo O’Higgins


* Región del Maule


* Región del Biobío


* Región de La Araucanía


* Región de Los Lagos


* Región Aisén del Gral.Carlos Ibáñez del Campo


* Región de Magallanes y de la Antártica Chilena


* Región Metropolitana de Santiago


* Región de Los Ríos


* Región de Arica y Parinacota


* Región de Ñuble

En conjunto a la data sobre las estaciones, extrajímos también los identificadores para cada territorio y los distribuidores de combustible para una estandarización más precisa en su posterior analisis y limpieza.

## **GET: ESTACIONES**

```python
import requests
import pandas as pd

# Importar la librería para manejar secretos
from google.colab import userdata

# 1) Configuramos la petición
url = "https://api.cne.cl/api/v4/estaciones"

token = userdata.get('CNE_API_TOKEN')


headers = {
    "Authorization": f"Bearer {token}"
}


# 2) Solicitud a la API
respuesta = requests.get(url, headers=headers)

if respuesta.status_code == 200:
    datos_json = respuesta.json()

    # Manejo de la estructura del JSON
    if isinstance(datos_json, dict) and 'data' in datos_json:
        lista_estaciones = datos_json['data']
    else:
        lista_estaciones = datos_json

    # 4) Aplanamos y creamos el DataFrame de manera que se pueda trabajar correctamente.
    df = pd.json_normalize(lista_estaciones)

    print("--- DATOS APLANADOS AUTOMÁTICAMENTE ---")
    # Mostramos las primeras columnas para verificar el aplanamiento
    print(df[['codigo', 'razon_social', 'distribuidor.marca', 'ubicacion.nombre_region', 'precios.93.precio']].head())

else:
    # Manejo en caso de error en la solicitud
    print(f"Error en la consulta. Código: {respuesta.status_code}")
    print(respuesta.text)
```

## **GET: TIPO COMBUSTIBLES**

```python
import requests
import pandas as pd

# 1) Configuración de la petición
url_tipos = "https://api.cne.cl/api/v4/combustible/vehicular/tiposcombustibles"

headers = {
    "Authorization": f"Bearer {token}"
}

# 2) Solicitud a la API
respuesta = requests.get(url_tipos, headers=headers)

# 3) Procesamiento y Normalización
if respuesta.status_code == 200:
    datos_json = respuesta.json()

    # Extraemos la lista de datos de forma segura
    if isinstance(datos_json, dict) and 'data' in datos_json:
        lista_tipos = datos_json['data']
    else:
        lista_tipos = datos_json

    # 4) Aplanamos y creamos el DataFrame de manera que se pueda trabajar correctamente.
    df_tipos_combustibles = pd.json_normalize(lista_tipos)

    print("--- DICCIONARIO DE TIPOS DE COMBUSTIBLES (NORMALIZADO) ---")
    print(df_tipos_combustibles.head(10))

else:
    print(f"Error en la consulta. Código: {respuesta.status_code}")
    print(respuesta.text)
```

## **GET: DISTRIBUIDORES**
