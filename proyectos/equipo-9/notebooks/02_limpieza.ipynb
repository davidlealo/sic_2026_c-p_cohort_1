```python
import pandas as pd

# 1. Agregamos 'season' a las columnas necesarias
columnas_necesarias = ['category', 'season', 'current_price', 'markdown_percentage', 'is_returned', 'customer_rating']

df = pd.read_csv('../data/dataset_original.csv', usecols=columnas_necesarias)

# 2. Limpieza de nulos
df.dropna(subset=['category', 'season', 'current_price', 'is_returned', 'customer_rating'], inplace=True)

# 3. Transformaciones
df['devuelto_num'] = df['is_returned'].replace({'Yes': 1, 'No': 0, True: 1, False: 0})
df['ingreso_neto'] = df.apply(lambda row: 0 if row['devuelto_num'] == 1 else row['current_price'], axis=1)

# 4. Exportar
df.to_csv('../data/dataset_limpio.csv', index=False)
print("Datos limpios actualizados con la columna temporada (season).")
```
