# Análisis Exploratorio de Datos (EDA)
**Objetivo:** Descomponer y entender la estructura del dataset original, evaluando la distribución de categorías, temporadas, tallas, fechas y niveles de stock antes de proceder con la limpieza.

```python
import pandas as pd
df = pd.read_csv('../data/dataset_original.csv')
print("Filas y columnas totales:", df.shape)
```

## Descomposición de variables clave
A continuación, analizaremos el volumen por categoría y temporada, la distribución de tallas (buscando nulos), el estado del inventario y el rango temporal de las ventas.

```python
# 1. Categorías y Temporadas
print("--- Volumen por Categoría ---")
display(df['category'].value_counts())
print("\n--- Volumen por Temporada ---")
display(df['season'].value_counts())

# 2. Análisis de Tallas
print("\n--- Distribución de Tallas (Incluyendo Nulos) ---")
display(df['size'].value_counts(dropna=False))

# 3. Comportamiento del Stock
print("\n--- Resumen Estadístico del Stock ---")
display(df['stock_quantity'].describe())
stock_critico = df[df['stock_quantity'] <= 0]
print(f"Transacciones con stock 0 o negativo: {len(stock_critico)}")

# 4. Rango Temporal
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
print("\n--- Rango de Fechas ---")
print("Primera transacción:", df['purchase_date'].min())
print("Última transacción:", df['purchase_date'].max())
```
