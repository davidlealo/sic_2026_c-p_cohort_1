```python
import pandas as pd

df = pd.read_csv('../data/dataset_limpio.csv')

# Agrupamos por categoría para ver el panorama general
resumen = df.groupby('category').agg({
    'ingreso_neto': 'sum',
    'devuelto_num': 'mean', # Esto nos dará la Tasa de Devolución
    'customer_rating': 'mean'
}).reset_index()

resumen['tasa_devolucion_%'] = resumen['devuelto_num'] * 100

print("Resumen de Rentabilidad y Satisfacción por Categoría:")
display(resumen.sort_values(by='ingreso_neto', ascending=False))
```
